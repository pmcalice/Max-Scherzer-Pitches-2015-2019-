```python
import pandas as pd
early_pitches = pd.read_csv("res/pitches.csv")
drop = ["px",
"pz",
"end_speed",
"spin_rate",
"spin_dir",
"break_angle",
"break_length",
"break_y",
"ax",
"ay",
"sz_bot",
"sz_top",
"type_confidence",
"vx0",
"vy0",
"vz0",
"x",
"x0",
"y",
"y0",
"z0",
"pfx_x",
"pfx_z",
"nasty",
"zone",
"code",
"type",
"event_num",
"start_speed",
"az",]
early_pitches.drop(drop, axis = 1, inplace = True)
early_pitches.head()

early_atbats = pd.read_csv("res/atbats.csv")
drop = ["event", "g_id", "o", "p_throws"]
early_atbats.drop(drop, axis = 1, inplace = True)
early_atbats.head()

players = pd.read_csv("res/player_names.csv")
players["player"] = players["first_name"] + " " + players["last_name"]
players.drop(["first_name", "last_name"], axis = 1, inplace = True)
players.head()

def get_player(id):
    if id in players.values:
        ind = players[players['id'] == id].index[0]
        return players.loc[ind, "player"]
    else:
        return "NOPE"

early_atbats["pitcher_name"] = early_atbats["pitcher_id"].map(get_player)
early_atbats = early_atbats[early_atbats['pitcher_name'] == 'Max Scherzer']
early_atbats["batter_name"] = early_atbats["batter_id"].map(get_player)
early_atbats.head()

df2015to2018 = pd.merge(early_atbats, early_pitches, on = "ab_id", how = "left")
df2015to2018.head()

pitches_2019 = pd.read_csv("res/2019_pitches.csv")
drop = ["px",
"pz",
"end_speed",
"spin_rate",
"spin_dir",
"break_angle",
"break_length",
"break_y",
"ax",
"ay",
"sz_bot",
"sz_top",
"type_confidence",
"vx0",
"vy0",
"vz0",
"x",
"x0",
"y",
"y0",
"z0",
"pfx_x",
"pfx_z",
"nasty",
"zone",
"code",
"type",
"event_num",
"start_speed",
"az",]
pitches_2019.drop(drop, axis = 1, inplace = True)
pitches_2019.head()

atbats_2019 = pd.read_csv("res/2019_atbats.csv")
drop = ["event", "g_id", "o", "p_throws"]
atbats_2019.drop(drop, axis = 1, inplace = True)
atbats_2019.head()

atbats_2019["pitcher_name"] = atbats_2019["pitcher_id"].map(get_player)
atbats_2019 = atbats_2019[atbats_2019['pitcher_name'] == 'Max Scherzer']
atbats_2019["batter_name"] = atbats_2019["batter_id"].map(get_player)
atbats_2019.head()

df2019 = pd.merge(atbats_2019, pitches_2019, on = "ab_id", how = "left")
df2019.head()

all_pitches = pd.concat([df2015to2018, df2019])
all_pitches.head()

hitters = pd.read_csv("res\hitters.csv")
hitters.head()

hitters["batter_name"] = hitters["Name"]
hitters.drop(["Contact%", "Name"], axis = 1, inplace = True)

df = pd.merge(all_pitches, hitters, on = "batter_name", how = "left")
df

df["pitch_thrown"] = df["pitch_type"]
drop = ["ab_id", "batter_id", "pitcher_id", "pitcher_name", "batter_name", "Team", "playerid", "pitch_type", "Z-Contact%"]
df = df.drop(drop, axis = 1)
df.head()

df

df.to_csv("max_scherzer_pitchers_2015_to_2019")
```
